5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

In contrast, the CDF represents the cumulative probability that a random variable takes on a value less than or equal to a specific value, summing up the probabilities from the PDF. Essentially, while the PDF provides the probability density for individual values, the CDF shows the cumulative probability up to a certain value. Both functions are essential for understanding the distribution and behavior of random variables.

13.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?
   - The Central Limit Theorem (CLT) is a fundamental principle in statistics that states that the distribution of the sample mean of a sufficiently large number of independent, identically distributed random variables will approximate a normal distribution, regardless of the original distribution of the variables. This theorem is critical because it allows statisticians to make inferences about population parameters using sample data. It underpins many statistical methods, including hypothesis testing and confidence intervals, by enabling the use of the normal distribution as an approximation for the sampling distribution of the sample mean. The CLT provides a foundation for making reliable and accurate predictions based on sample data, even when the underlying population distribution is unknown.

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
# Ans

import numpy as np
import matplotlib.pyplot as plt
def calculate_z_scores(data):
    """
    Calculate Z-scores from a dataset.

    Parameters:
    data (array-like): The dataset

    Returns:
    array-like: Z-scores of the dataset
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores
def plot_z_scores(z_scores):
    """
    Plot Z-scores.

    Parameters:
    z_scores (array-like): Z-scores to plot
    """
    plt.hist(z_scores, bins=30, density=True, edgecolor='black')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.title('Histogram of Z-scores')
    plt.show()
data = np.random.normal(0, 1, 1000)
z_scores = calculate_z_scores(data)
plot_z_scores(z_scores)
```

19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
# Ans

import numpy as np
from scipy import stats
def hypothesis_testing(sample_data, population_mean, alpha=0.05):
    """
    Perform hypothesis testing using Z-statistics.
    Parameters:
    sample_data (array-like): The sample data
    population_mean (float): The population mean
    alpha (float): Significance level (default is 0.05)
    Returns:
    float: Z-score
    float: p-value
    bool: Whether to reject the null hypothesis
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    reject_null = p_value < alpha
    return z_score, p_value, reject_null
sample_data = np.random.normal(50, 10, 30)
population_mean = 55
alpha = 0.05
z_score, p_value, reject_null = hypothesis_testing(sample_data, population_mean, alpha)
print("Z-score:", z_score)
print("p-value:", p_value)
print("Reject the null hypothesis:", reject_null)
```

20.  Create a confidence interval for a dataset using Python and interpret the result.

```python
# Ans

import numpy as np
import scipy.stats as stats
def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.
    Parameters:
    data (array-like): The dataset
    confidence (float): The confidence level (default is 0.95)
    Returns:
    tuple: Lower and upper bounds of the confidence interval
    """
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return mean - margin_of_error, mean + margin_of_error
data = np.random.normal(50, 10, 30)
confidence = 0.95
lower_bound, upper_bound = confidence_interval(data, confidence)
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
# Ans

import numpy as np
import matplotlib.pyplot as plt
def plot_normal_pdf(mean, std_dev):
    """
    Calculate and plot the PDF of a normal distribution.
    Parameters:
    mean (float): Mean of the distribution
    std_dev (float): Standard deviation of the distribution
    """
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
    y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
    plt.plot(x, y, label='Normal Distribution PDF')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title('Probability Density Function (PDF) of Normal Distribution')
    plt.legend()
    plt.show()
mean = 0
std_dev = 1
plot_normal_pdf(mean, std_dev)
```

23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
# Ans

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
def plot_poisson_cdf(lambda_, size):
    """
    Calculate and plot the CDF of a Poisson distribution.
    Parameters:
    lambda_ (float): Average rate (Œª)
    size (int): Number of samples
    """
    samples = np.random.poisson(lambda_, size)
    values, counts = np.unique(samples, return_counts=True)
    cdf = np.cumsum(counts) / size
    plt.step(values, cdf, where='mid', label='Poisson CDF')
    plt.xlabel('Number of Events')
    plt.ylabel('Cumulative Probability')
    plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.legend()
    plt.show()
lambda_ = 5
size = 1000
plot_poisson_cdf(lambda_, size)
```

```python
# Ans

import numpy as np
import matplotlib.pyplot as plt
def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.
    Parameters:
    data1 (array-like): The first dataset
    data2 (array-like): The second dataset
    """
    std_dev1 = np.std(data1, ddof=1)
    std_dev2 = np.std(data2, ddof=1)
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")
    plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1', edgecolor='black')
    plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2', edgecolor='black')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Comparison of Standard Deviations')
    plt.legend()
    plt.show()
data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(60, 15, 1000)
compare_standard_deviations(data1, data2)
```

26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
# Ans

import numpy as np
mean = 50
std_dev = 10
size = 1000
data = np.random.normal(mean, std_dev, size)
data_range = np.ptp(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

27.  Implement Z-score normalization on a dataset and visualize its transformation.

```python
# Ans

import numpy as np
import matplotlib.pyplot as plt
def z_score_normalization(data):
    """
    Perform Z-score normalization on a dataset.
    Parameters:
    data (array-like): The dataset
    Returns:
    array-like: Normalized dataset
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data
def plot_data(original_data, normalized_data):
    """
    Plot the original and normalized datasets.
    Parameters:
    original_data (array-like): The original dataset
    normalized_data (array-like): The normalized dataset
    """
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=30, edgecolor='black')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Original Data')
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, edgecolor='black')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.title('Z-score Normalized Data')
    plt.tight_layout()
    plt.show()
data = np.random.normal(50, 10, 1000)
normalized_data = z_score_normalization(data)
plot_data(data, normalized_data)
```

28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
# Ans

import numpy as np
import scipy.stats as stats
def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.
    Parameters:
    data (array-like): The dataset
    Returns:
    tuple: Skewness and kurtosis of the dataset
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis
data = np.random.normal(0, 1, 1000)
skewness, kurtosis = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness:.2f}")
```
